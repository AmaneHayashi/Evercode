```python
import numpy as np
from keras.callbacks import TensorBoard
from keras.utils import np_utils
from keras.layers import Dense, Dropout
from keras.models import Sequential
from keras.optimizers import SGD

LOG_DIR = './log'

MNIST_FILE = './mnist.npz'

def clear_log(log_dir = LOG_DIR):
    import shutil
    import os
    shutil.rmtree(log_dir)
    os.mkdir(log_dir)

clear_log()

with np.load(MNIST_FILE) as f:
    x_train, y_train = f['x_train'], f['y_train']
    x_test, y_test = f['x_test'], f['y_test']
    
x_train = x_train.reshape([-1, 784]) / 255
x_test = x_test.reshape([-1, 784]) / 255
y_train = np_utils.to_categorical(y_train, num_classes = 10)
y_test = np_utils.to_categorical(y_test, num_classes = 10)

model = Sequential()

model.add(Dense(512, activation='relu', input_shape=[784], name='fc1'))

model.add(Dropout(0.5, name='dropout'))

model.add(Dense(256, activation='relu', name='fc2'))
model.add(Dense(128, activation='relu', name='fc3'))

model.add(Dense(10, activation='softmax', name='fc4'))

model.summary()

model.compile(optimizer=SGD(lr=0.005), loss='categorical_crossentropy',
              metrics=['accuracy'])

callback = TensorBoard(log_dir=LOG_DIR, histogram_freq = 1, write_grads=True)

model.fit(x_train, y_train, batch_size = 64, epochs = 10, 
          validation_split = 0.1, callbacks=[callback])

loss, accuracy = model.evaluate(x_test, y_test)
print('loss: %f\naccuracy : %f' %(loss, accuracy))
```
